```python
import pandas as pd
import json
import json_lines
```

```python
obj = json.load(open("nendai_ryuko/album_sales_processed.json"))
```

```python
target_artist = obj[0]["artist"]
target_title = obj[0]["album_data"][1]["title"]
print(target_artist, target_title)
```

```python
lyric_dict_utaten = {}
with json_lines.open("utaten/lyric.jl") as reader:
    for lyric in reader:
        artist = lyric["artist"]
        title = lyric["title"]
        lyric_body = lyric["lyric_body"]
        lyric_dict_utaten[(artist, title)] = lyric_body
len(lyric_dict_utaten)
```

```python
def is_equal(a, b):
    if a[0:5] == b[0:5]:
        return True
    if a[-5:] == b[-5:]:
        return True
    return False

def search_lyric(lyric_dict_utaten, target_artist, target_title):
    if (target_artist, target_title) in lyric_dict_utaten.keys():
        target_lyric = lyric_dict_utaten[(target_artist, target_title)]
        return target_title, target_lyric
    for artist, title in lyric_dict_utaten.keys():
        if is_equal(title, target_title):
            target_lyric = lyric_dict_utaten[(artist, title)]
            return title, target_lyric
    return None, None
```

```python
search_lyric(lyric_dict_utaten, target_artist, target_title)
```

```python
lyric_dict ={} 
counter_all = 0
counter = 0
for title_list in obj:
    artist_name = title_list["artist"]
    lyric_dict_by_one_artist = {}
    for lyric_data in title_list["album_data"]:
        title_name = lyric_data["title"]
        counter_all += 1
        if counter_all % 100 == 0: print("counter: {}, {}".format(counter_all, counter));
        title_true, lyric = search_lyric(lyric_dict_utaten, artist_name, title_name)
        if lyric is not None:
            counter += 1
            lyric_dict_by_one_artist[title_name] = {"true_title": title_true, "lyric_body": lyric}
            continue
#         else:
#             print(artist_name, title_name)
    lyric_dict[artist_name] = lyric_dict_by_one_artist
print("counter: {}, {}".format(counter_all, counter));
```
